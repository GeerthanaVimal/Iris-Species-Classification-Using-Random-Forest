```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
#creating an object called iris with iris data
iris=load_iris()
print(iris)
```

```python
#creating a dataframe with four feature variables
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
#viewing the top 5 rows
df.head()
```

```python
#adding a new column for a species name
df['species']=pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
#creating test and train data
df['is_train']=np.random.uniform(0,1,len(df)) <=0.75
df.head()
```

```python
#creating dataframes with test rows and training rows
train, test=df[df['is_train']==True], df[df['is_train']==False]
#show the number of observationsfor the test and training dataframes
print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:', len(test))
```

```python
#create a list of the feature column's name
features=df.columns[:4]
#view features
features
```

```python
#converting each species name into digits
y=pd.factorize(train['species'])[0]
#viewing target
y
```

```python
#creating a random forest
clf=RandomForestClassifier(n_jobs=2, random_state=0)
#training the classifier
clf.fit(train[features],y)
```

```python
#applying the trained classifier to the test
#test[features]
clf.predict(test[features])
```

```python
#viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
#mapping names for the plants for each predicted plant class
preds=iris.target_names[clf.predict(test[features])]
#view the PREDICTED species for the first five observations
preds[0:5]
```

```python
#view the ACTUAL species for the first five observations
test['species'].head()
```

```python
#creating confusion matrix
pd.crosstab(test['species'],preds,rownames=['Actual Species'], colnames=['Predicted species'])
```
